# Problem 1: Fitting the Spectrum of an Astronomical Source

The spectrum of an astronomical source is defined to be the power (energy/time) it emits per unit wavelength interval
$$
L_\lambda  = \left({\rm \frac{energy}{time \cdot wavelength}}\right); \quad \text{units: }\left[\frac{\rm J}{{\rm s \cdot \AA}}\right]\text{},
$$
where $\lambda$ is the wavelength and $1{\rm \AA} = 10^{-10}~{\rm m}$ is the Angstrom unit of length. If the source is located at some distance $d$ from the Earth, we will observe an energy flux
$$
f_{\lambda} = \frac{L_{\lambda}}{4\pi d^2} = \left({\rm \frac{energy}{time \cdot area \cdot wavelength}}\right);  \quad \text{units: }\left[\frac{\rm J}{{\rm s \cdot m^2 \cdot \AA}}\right]\text{}.  
$$
(we are ignoring effects that are important at cosmological distances due to the expansion of the universe). For a telescope with an aperture with area $A$ (the area of its mirror or lens) the number of photons of wavelength $\lambda$ that will be detected in a time interval $\Delta t$ and a wavelength interval $\Delta \lambda$ is given by
$$
N_\lambda = \frac{f_{\lambda} A \Delta t \Delta \lambda}{h \frac{c}{\lambda}} = \left({\rm \frac{energy}{photon~energy}}\right) \quad  \text{units: [number of photons]}, 
$$
where $c$ is the speed of light, $h$ is Planck's constant, and $\nu = c/\lambda$ is the frequency of the photons. Here $\Delta t$ would be the exposure time used for the digial camera used to record the spectrum,  and $\Delta \lambda$ the wavelength interval spanned by a single pixel on that cameras detector. 

The number of photons $N_\lambda$ will obey Poisson statistics, that is for any spectral bin centered on $\lambda$ (i.e. corresponding to a pixel on the detector)
$$
P(k | N_\lambda) = \frac{N_\lambda^k}{k!} e^{-N_\lambda},
$$
where $k$ is the discrete integer number of photons detected in that pixel. In other words the number of photons measured per pixel
fluctuates from the average rate implied by the energy flux, $f_{\lambda}$, of the source because the arrival of photons at our detector  constitutes a *counting process* that follows the Poisson distribution.

In this problem we will consider the spectrum of a hypothetical galaxy which has a spectrum that is described by a power-law continuum with a superimposed emission line at a specific known wavelength
$$
N_\lambda = N_{\rm cont}\left(\frac{\lambda}{1200~{\rm \AA}}\right)^{\alpha} + N_{\rm line} \mathcal{N}(\lambda |\lambda_{\rm line}, \sigma_{\rm line}),
$$
where $N_{\rm cont}$ is the continuum amplitude, $\alpha$ is the power law index,  $\mathcal{N}(\lambda |\lambda_{\rm line}, \sigma_{\rm line})$ is a Gaussian emission line profile centered on the wavelength $\lambda_{\rm line}=1216~{\rm \AA}$ with standard deviation $\sigma_{\rm line} = 17~{\rm \AA}$, and $N_{\rm line}$ is the total  number of photons resulting from the emission line (i.e. the Gaussian emission line profile is normalized so its integral over wavelength is unity). The unknown model parameters are thus $\boldsymbol{\theta} = (N_{\rm cont}, \alpha, N_{\rm line})$. **Note the domain of these model parameters -- both $N_{\rm cont} \in [0,\infty]$ and $N_{\rm line} \in [0,\infty]$, but $\alpha$ can take on any real value.**

Visualizing a Poisson process in the low count regime requires some thought since the process is not Gaussian and so error bars corresponding to e.g. the 16th and 84th percentile credibility interval are generally asymmetric, and only become symmetric in the limit of large counts (where the Poisson distribution tends toward Gaussianity). Furthermore,  if we detect zero photons in a pixel $k=0$ --- a common occurrence if the rate $N_\lambda$ is small but non-zero -- then it is unclear how to represent the error bar on such a measurement since 
$$
\lim_{N_\lambda \to 0} P(k | N_\lambda) = \begin{cases} 
1 & \text{for } k=0\\
0 & \text{otherwise},  
\end{cases}
$$
so there is no credible interval that we can compute on our measured value of $k$. But zero photons does not necessarily mean that the true rate $N_\lambda=0$, it could be that we just didn't detect any photons in that pixel by chance. 

This is something that scientists actually spend a lot of time debating. You can read about the various options that have been discussed [here](https://docs.astropy.org/en/stable/api/astropy.stats.poisson_conf_interval.html). For the purposes of the current problem we will adopt the following convention, the credibility interval, ${\rm CI}$, which corresponds to the upper and lower error bars that we will plot are defined by 
$$
{\rm CI} = (k + 0.5 - \sqrt{k + 0.25}, k + 0.5 + \sqrt{k + 0.25})
$$
which is the `pearson` option discussed here [here](https://docs.astropy.org/en/stable/api/astropy.stats.poisson_conf_interval.html). 

Note these somewhat arbitrary choices for the error bars **only impact our ability to visualize the data**. As statistically sophisticated physicists, we will need to construct the correct likelihood function which is the full generative probabilistic model of our data. 

Below I provide you with a plotting function that will make a nice plot of the spectrum with error bars following a reasonable convention.

```python
def plot_spectrum(ax, wave, N_lam_obs): 
    """
    Utility function to plot a spectrum in the low photon regime. 
    
    Parameters
    ----------
        ax: matplotlib axis object
            The axis to plot the spectrum on. This is a matplotlib axes object created by i.e. 
            the object oriented API of matplotlib. 
        wave: array-like
            The wavelength array of the spectrum.
        N_lam_obs: array-like
            The observed number of photons detected per wavelength bin.
    
    """

    # Calculate bin midpoints
    bin_widths = np.diff(wave) / 2.0
    bin_widths = np.append(bin_widths[0], bin_widths)*0.97
    N_lam_lower = N_lam_obs + 0.5 - np.sqrt(N_lam_obs + 0.25)
    N_lam_upper = N_lam_obs + 0.5 + np.sqrt(N_lam_obs + 0.25)
 
    for wv, bin_width, Nval, Nval_lo, Nval_hi in zip(wave, bin_widths, N_lam_obs, N_lam_lower, N_lam_upper):
        # Draw rectangles for the error bars
        ax.fill_betweenx([Nval_lo, Nval_hi], wv-bin_width, wv+bin_width, color='dodgerblue', alpha=0.5)
        # Draw horizontal lines for the measureed value
        ax.hlines(Nval, wv-bin_width, wv+bin_width, color='black', linestyle='-')
    
```
