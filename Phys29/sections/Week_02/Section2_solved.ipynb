## Problem 3: Simple Debugging

The following two cells are two implementations to calculate the classical Coulomb force between a proton and an electron at the Bohr radius:
\begin{equation}
F_{Coulomb} = \frac{kq_{1}q_{2}}{r^2}

Where 

$k = 8.98 \times 10^{9} \; Nm^2C^{-2}$ is the Coulomb constant

$q_1 = -q_2 = 1.602 \times 10^{-19} C$ is the elementary charge that proton and electrons carry

This implementation has mistakenly divided by r instead of r^2. Fixing this mistake provides the expected result.

#### Now, as an exercise, please add comments to the following piece of code. It is a piece of code that calculates the gravitational force between two objects given the masses of the two objects (M and m) and the distance between them (r), via the formula:
\begin{equation}
F = \frac{GMm}{r^2}
\end{equation}

```python
def calculate_gravitational_force(M, m, r):
    """
    Function to calculate the gravitational force between two objects. 

    Parameters
    ----------
    M : float
        Mass of the first object in kg.
    m : float
        Mass of the second object in kg.
    r : float
        Distance between the two objects in m.

    Returns
    -------
    F : float
        Gravitational force between the two objects in Newtons N = kg m/s^2.
    """

    # Gravitational Constant: N m kg**-2
    G = 6.67e-11

    # Calculate the gravitational force
    F = G*M*m / r**2

    # Return the gravitational force as a float value
    return F

mass_object1 = 1    # Set the Mass of object 1 to 1 kg
mass_object2 = 6e24     # Set the Mass of Object 2 to 6e24 kg
distance_between_objects = 6.4e6     # Set the distance between the objects to 6.4e6 m

# Calculate the gravitational force for the parameters we have set
gravitational_force = calculate_gravitational_force(mass_object1, mass_object2, distance_between_objects)

# Print out the gravitational force for our scenario rounded to to 2 digits
print('The gravitational force between the two objects is {:.2f} N.'.format(gravitational_force))
```

You’ll need the equation:
\begin{equation}
∆y = v_0t − \frac{1}{2}gt^2
\end{equation}
with $g ≈ 9.8 \;m s^{−2}$. 

To take a square root in Python, you can use the square root function (if you know how to call it), or you
can just raise an expression to the $\frac{1}{2}$ power. Remember order of operations when you do this!

What happens if you run your code with an initial velocity of $10 \;m s^{−1}$ (so that the ball does not have enough energy to reach a height of 15 m)?

```python
# your solution here

#Construct a function to easily call at any time step.
def delta_y(t,v0=20):
    """
    Function to calculate the change in height after some time, given some initial velocity. 

    Parameters
    ----------
    t : float
        time that has passed since the ball was thrown in s.
    v0 : float, optional
        Initial velocity of the thrown ball in m/s.

    Returns
    -------
    d_y : float
        Change in height since t=0 in m.
    """

    # Define the gravitational acceleration (m s**-2)
    g = 9.807

    # Calculate the change in height at time t
    d_y = v0 * t - 0.5 * g * t**2
    
    # Return the change in height in units of m
    return d_y
    


# Set initial height to 0 m
y_init = 0

# Set initial time to 0 s
t_init = 0

# Create variable to update with new time
t = t_init

# Create variable to update with new position
y = y_init + delta_y(t)

# While the height of the ball is less than 15 m, proceed through the loop
while y < 15:

    # Increase the time step FIRST, to ensure we have the right final time
    t += 0.0001

    # Increase the height of the ball according to the new time before checking the while condition
    y = y_init + delta_y(t)

# Print out the time value at which point the loop ended
print('The time which the ball first reached 15 m was at: {:.4f} s'.format(t))
```

The code never stopped running. This happened because the while loop is UNSAFE in the sense that there is no guarantee the condition will be met. In the situation that this is the case, such as when v0 = 10, then the while loop will run infinitely.

When you use a while loop, you want to make sure that either 1) You know EXACTLY when the loop will end or 2) you have some way of determining if the while loop is working properly. This can be seen in the next example.

(Please be aware you can shorten the output length by clicking on the left side of where the output is)

```python
# your solution here

#Construct a function to easily call at any time step.
def delta_y(t,v0=20):
    """
    Function to calculate the change in height after some time, given some initial velocity. 

    Parameters
    ----------
    t : float
        time that has passed since the ball was thrown in s.
    v0 : float, optional
        Initial velocity of the thrown ball in m/s.

    Returns
    -------
    d_y : float
        Change in height since t=0 in m.
    """

    # Define the gravitational acceleration (m s**-2)
    g = 9.807

    # Calculate the change in height at time t
    d_y = v0 * t - 0.5 * g * t**2
    
    # Return the change in height in units of m
    return d_y
    


# Set initial height to 0 m
y_init = 0

# Set initial time to 0 s
t_init = 0

# Create variable to update with new time
t = t_init



# NOW; Set v0 = 10 m s**-1
v0 = 10


# Create variable to update with new position
y = y_init + delta_y(t)

# Create a count variable to keep track of the while loop
count = 0

# While the height of the ball is less than 15 m, proceed through the loop
while y < 15:

    # Increase the time step FIRST, to ensure we have the right final time
    t += 0.001

    # Increase the height of the ball according to the new time before checking the while condition
    y = y_init + delta_y(t,v0)

    # If a certain number of steps have occurred, then print out the current time and height (Using the modulus function)
    if count%100 == 0:
        print(f"Step {count}: Time = {t};    Height = {y}")

    # Increment the count by one
    count += 1

# Print out the time value at which point the loop ended
print(t)
```

Clearly, we see that with the way the loop is currently structured, it will NEVER end. Having an output such as this is useful when we are unsure of how certain code is behaving.
